```python

# Filename for data with Isochrone fit for 1 Gyr
# These files are located in the folder IsochroneData
filename1="C:/Users/ellen/OneDrive/Desktop/U of A/spring 25/astr400b/Isochrone1.txt"
filename2="C:/Users/ellen/OneDrive/Desktop/U of A/spring 25/astr400b/Isochrone2.txt"
filename3="C:/Users/ellen/OneDrive/Desktop/U of A/spring 25/astr400b/Isochrone3.txt"
filename4="C:/Users/ellen/OneDrive/Desktop/U of A/spring 25/astr400b/Isochrone4.txt"
filename5="C:/Users/ellen/OneDrive/Desktop/U of A/spring 25/astr400b/Isochrone5.txt"
filename10="C:/Users/ellen/OneDrive/Desktop/U of A/spring 25/astr400b/Isochrone10.txt"
filename11="C:/Users/ellen/OneDrive/Desktop/U of A/spring 25/astr400b/Isochrone11.txt"
filename7="C:/Users/ellen/OneDrive/Desktop/U of A/spring 25/astr400b/Isochrone7.txt"
```

```python
# READ IN DATA
# "dtype=None" means line is split using white spaces
# "skip_header=8"  skipping the first 8 lines 
# the flag "names=True" creates arrays to store the date
#       with the column headers given in line 8 

# Read in data for an isochrone corresponding to 1 Gyr
data1 = np.genfromtxt(filename1,dtype=None,
                      names=True,skip_header=8)
```

```python
data10 = np.genfromtxt(filename10,dtype=None,
                      names=True,skip_header=8)
data11 = np.genfromtxt(filename11,dtype=None,
                      names=True,skip_header=8)
data5 = np.genfromtxt(filename5,dtype=None,
                      names=True,skip_header=8)
data4 = np.genfromtxt(filename4,dtype=None,
                      names=True,skip_header=8)
data3 = np.genfromtxt(filename3,dtype=None,
                      names=True,skip_header=8)
data7 = np.genfromtxt(filename7,dtype=None,
                      names=True,skip_header=8)
data2 = np.genfromtxt(filename2,dtype=None,
                      names=True,skip_header=8)
```

```python
# Plot Isochrones 
# For Carina

fig = plt.figure(figsize=(10,10))
ax = plt.subplot(111)

# Plot Isochrones

# Isochrone for 1 Gyr
# Plotting Color vs. Difference in Color 
plt.plot(data1['B']-data1['R'], data1['R'], color='blue', 
         linewidth=5, label='1 Gyr')
'''
plt.plot(data10['B']-data10['R'], data10['R'], color='red', 
         linewidth=5, label='10 Gyr')
'''
plt.plot(data11['B']-data11['R'], data11['R'], color='green', 
         linewidth=5, label='11 Gyr')
'''
plt.plot(data3['B']-data3['R'], data3['R'], color='yellow', 
         linewidth=5, label='3 Gyr')
plt.plot(data4['B']-data4['R'], data4['R'], color='purple', 
         linewidth=5, label='4 Gyr')
plt.plot(data5['B']-data5['R'], data5['R'], color='orange', 
         linewidth=5, label='5 Gyr')
'''
plt.plot(data7['B']-data7['R'], data7['R'], color='deeppink', 
         linewidth=5, label='7 Gyr')
plt.plot(data2['B']-data2['R'], data2['R'], color='orange', 
         linewidth=5, label='2 Gyr')

# Add axis labels
plt.xlabel('B-R', fontsize=22)
plt.ylabel('M$_R$', fontsize=22)

#set axis limits
plt.xlim(-0.5,2)
plt.ylim(5,-2.5)

#adjust tick label font size
label_size = 22
matplotlib.rcParams['xtick.labelsize'] = label_size 
matplotlib.rcParams['ytick.labelsize'] = label_size

# add a legend with some customizations.
legend = ax.legend(loc='upper left',fontsize='x-large')

#add figure text
plt.figtext(0.5, 0.15, 'CMD for Carina dSph', fontsize=22)

plt.savefig('IsochroneCarina.png')
```
